```python
import pandas as pd
data = pd.read_csv('Sub_Oil_VLCC_Monthly.csv')
data
```

```python
#Splitting the data into training and testing data

X = data[:-1]  #whole data exclude the last row as X

#target columns
good_result1_target_column = data['542236']
good_result2_target_column = data['67321']
mid_result_target_column = data['549295']
bad_result1_target_column = data['41108']
bad_result2_target_column = data['541982']

#target column exclude the first row as y
good1_y = good_result1_target_column[1:]
good2_y = good_result2_target_column[1:] 
mid_y = mid_result_target_column[1:] 
bad1_y = bad_result1_target_column[1:]
bad2_y = bad_result2_target_column[1:]

n = len(X)-36 #number of data points for training and testing

X_train = X.iloc[:n, :] #training data
X_test = X.iloc[n:, :] #testing data

#good1 target column training and testing data
good1_y_train = good1_y.iloc[:n]
good1_y_test = good1_y.iloc[n:]

#good2 target column training and testing data
good2_y_train = good2_y.iloc[:n]
good2_y_test = good2_y.iloc[n:]

#mid target column training and testing data
mid_y_train = mid_y.iloc[:n]
mid_y_test = mid_y.iloc[n:]

#bad1 target column training and testing data
bad1_y_train = bad1_y.iloc[:n]
bad1_y_test = bad1_y.iloc[n:]

#bad2 target column training and testing data
bad2_y_train = bad2_y.iloc[:n]
bad2_y_test = bad2_y.iloc[n:]

#print(X_train)
#print(X_test)
#print(good1_y_train)
#print(good1_y_test)
#print(good2_y_train)
#print(good2_y_test)
#print(mid_y_train)
#print(mid_y_test)
#print(bad1_y_train)
#print(bad1_y_test)
#print(bad2_y_train)
print(bad2_y_test)
```

```python
#Fitting the model, using DecisionTreeRegressor
from sklearn.tree import DecisionTreeRegressor

model1 = DecisionTreeRegressor(random_state=42)
model1.fit(X_train, good1_y_train)

model2 = DecisionTreeRegressor(random_state=42)
model2.fit(X_train, good2_y_train)

model3 = DecisionTreeRegressor(random_state=42)
model3.fit(X_train, mid_y_train)

model4 = DecisionTreeRegressor(random_state=42)
model4.fit(X_train, bad1_y_train)

model5 = DecisionTreeRegressor(random_state=42)
model5.fit(X_train, bad2_y_train)

#Predicting the model
good1_y_pred = model1.predict(X_test)
good2_y_pred = model2.predict(X_test)
mid_y_pred = model3.predict(X_test)
bad1_y_pred = model4.predict(X_test)
bad2_y_pred = model5.predict(X_test)

print(good1_y_pred)
print(good2_y_pred)
print(mid_y_pred)
print(bad1_y_pred)
print(bad2_y_pred)
```

```python
#Calculating the accuracy of the model
#Loop through the predicted values and the actual values and calculate the difference
#Then calculate the percentage difference
good1_accuracy = []
good2_accuracy = []
mid_accuracy = []
bad1_accuracy = []
bad2_accuracy = []
for i in range(len(good1_y_pred)):
    good1_accuracy.append(round((1-(abs(good1_y_pred[i] - good1_y_test.iloc[i])/good1_y_test.iloc[i]))*100,2))
    good2_accuracy.append(round((1-(abs(good2_y_pred[i] - good2_y_test.iloc[i])/good2_y_test.iloc[i]))*100,2))
    mid_accuracy.append(round((1-(abs(mid_y_pred[i] - mid_y_test.iloc[i])/mid_y_test.iloc[i]))*100,2))
    bad1_accuracy.append(round((1-(abs(bad1_y_pred[i] - bad1_y_test.iloc[i])/bad1_y_test.iloc[i]))*100,2))
    bad2_accuracy.append(round((1-(abs(bad2_y_pred[i] - bad2_y_test.iloc[i])/bad2_y_test.iloc[i]))*100,2))

print(good1_accuracy)
print(good2_accuracy)
print(mid_accuracy)
print(bad1_accuracy)
print(bad2_accuracy)

#Calculate the average accuracy in 2 decimal places
good1_avg_accuracy = round(sum(good1_accuracy) / len(good1_accuracy), 2)
good2_avg_accuracy = round(sum(good2_accuracy) / len(good2_accuracy), 2)
mid_avg_accuracy = round(sum(mid_accuracy) / len(mid_accuracy), 2)
bad1_avg_accuracy = round(sum(bad1_accuracy) / len(bad1_accuracy), 2)
bad2_avg_accuracy = round(sum(bad2_accuracy) / len(bad2_accuracy), 2)

print(good1_avg_accuracy)
print(good2_avg_accuracy)
print(mid_avg_accuracy)
print(bad1_avg_accuracy)
print(bad2_avg_accuracy)
```
