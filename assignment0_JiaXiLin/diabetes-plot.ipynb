```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
diabetes_dataset = pd.read_csv("diabetes.csv")
print(diabetes_dataset.shape)
```

```python
print(diabetes_dataset.head())
```

```python
diabetes_columns = diabetes_dataset[['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age', 'Outcome']]
diabetes_columns['Outcome'] = diabetes_columns['Outcome'].replace([0, 1], ['No', 'Yes'])
diabetes_colours = {'No': 'red', 'Yes': 'green'}


# hue specifies which column you want to change colours of, then palette makes it possible to change colours using like a dictionary
sns.pairplot(diabetes_columns, hue='Outcome', palette=diabetes_colours)

plt.show()
```
